```python
import sys
sys.path.insert(0, '../')
import numpy as np
from lib_TMG import Visualization
from lib_TMG import TMGSE as se
```

```python
# Definition of the base tensors

# Center and background eigenvalues
evals_c = np.array([0.4, 0.4, 0.4])
# Eigenvalues in the "axes"
evals_base = np.array([1.0, 0.3, 0.3])

# Eigenvectors in the x axis
evecs_x = np.array([[1,0,0],[0,1,0],[0,0,1]])
# Eigenvectors in the y axis
evecs_y = np.array([[0,1,0],[1,0,0],[0,0,1]])
# Eigenvectors in the z axis
evecs_z = np.array([[0,0,1],[0,1,0],[1,0,0]])

# Generating data in the correct format
evals = np.ones((3,3,3,3))
evecs = np.ones((3,3,3,3,3))
# Assign background eigenvalues to all voxels
evals = evals*evals_c
# Assign x axis eigenvectors to all voxels (it doesn't matter for the spherical tensorss)
evecs = evecs*evecs_x

# Assigns the correct eigenvalues to the axes tensors
evals[1,1,0] = evals[1,0,1] = evals[0,1,1] = evals[2,1,1] = evals[1,2,1] = evals[1,1,2] = evals_base
# Assigns the correct eigenvectors to the remaining axes
evecs[1,1,0] = evecs[1,1,2] = evecs_z
evecs[1,0,1] = evecs[1,2,1] = evecs_y

# Visualization of the resulting tensors
Visualization.volumeTensor(evals, evecs)
```

```python
# Creates a fake FA image to visualize the tensors of the neighborhood
def fa_for_vis(shape, indices):
    
    # The fake FA image is used to make tensors outside the neighborhood darker
    fa_vis = np.ones(np.prod(shape))*0.2
    fa_vis[indices[1,1,1][:]] = 1
    fa_vis = np.reshape(fa_vis, (3,3,3))

    return fa_vis
```

```python
# 2-connected neighborhood

# Select the axis (x, y or z)
nbh_info = 'x'
# Getting the indices of the neighbors
indices = se.two_connected(evecs.shape, nbh_info)
# Creates a fake FA image
fa_vis = fa_for_vis(evals.shape[0:3], indices)

# Tensors of the neighborhood are brighter while tensors outside of it are darker
Visualization.volumeTensor(evals, evecs, fa=fa_vis)
```

```python
# 4-connected neighborhood

# Select the plane (xy, xz or yz)
nbh_info = 'xy'
# Getting the indices of the neighbors
indices = se.four_connected(evecs.shape, nbh_info)
# Creates a fake FA image
fa_vis = fa_for_vis(evals.shape[0:3], indices)

# Tensors of the neighborhood are brighter while tensors outside of it are darker
Visualization.volumeTensor(evals, evecs, fa=fa_vis)
```

```python
# 6-connected neighborhood (3D)

# Getting the indices of the neighbors
indices = se.six_connected(evecs.shape, None)
# Creates a fake FA image
fa_vis = fa_for_vis(evals.shape[0:3], indices)

# Tensors of the neighborhood are brighter while tensors outside of it are darker
Visualization.volumeTensor(evals, evecs, fa=fa_vis)
```
